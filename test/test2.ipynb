```python
import myfunctions as fun
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
#make the datasets and resizing the varaibles

x, y = make_regression(n_samples=1000, n_features=2, noise=10)
y = y.reshape(y.shape[0],1)
X = np.hstack((x, np.ones((x.shape[0],1))))
tetha = np.random.randn(X.shape[1], 1)
```

```python
#show graphe before training 
plt.scatter(X[:,0], y)
plt.title('Representation colomun 0 in 2D')
```

```python
plt.scatter(X[:,1], y)
plt.title('Representation colomun 1 in 2D')
```

```python
#traning
tetha_final, cost_History = fun.gradient_Desc(X, y, tetha, 0.01, 500)

prediction = fun.modele(X, tetha_final)
R = fun.coef_Determination(y, prediction)

plt.plot(range(500), cost_History)
plt.title('Graphe for evolution of cost function')
R
```
